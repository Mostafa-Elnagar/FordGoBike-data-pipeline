# Data Exploration

```python
import pandas as pd
import numpy as np
import os
```

```python

dataframes = {}

for file in os.listdir('../data/extracted'):
    period = file.split('-')[0]
    df = pd.read_csv(f'../data/extracted/{file}')
    df['period'] = period
    dataframes[period] = df

df = pd.concat(dataframes.values())

df.info()
```

```python
stations = (
    df.groupby('start_station_id')
    .agg({'start_station_name': 'first', 'start_station_latitude': 'first', 'start_station_longitude': 'first'})
    .reset_index()
    .rename(columns={'start_station_id': 'station_id', 'start_station_name': 'station_name', 'start_station_latitude': 'latitude', 'start_station_longitude': 'longitude'})
)
stations.info()
```

```python
pd.merge(
    df[df["start_station_id"].isna()],
    stations,
    left_on=['start_station_latitude', 'start_station_longitude'],
    right_on=['latitude', 'longitude'],
    how='left'
)
```

```python
def nearest_station(lat, long):
    stats = stations.copy()
    stats["dist"] = ((stats[["latitude", "longitude"]] - np.array([lat, long])) ** 2).sum(axis=1)
    return stats.iloc[stats["dist"].idxmin()]
```
